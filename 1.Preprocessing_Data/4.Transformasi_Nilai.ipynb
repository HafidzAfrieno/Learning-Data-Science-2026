```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../0.dataset/dataset_penjualan_emas_kotor.csv')
df.head(10)
```

```python
colom_numerik = df.select_dtypes(include=np.number).columns

skewsness = df.select_dtypes(include=np.number).skew()
colom_normals = skewsness[skewsness.abs() <= 0.5].index
```

## 1.Normalisasi Data

### 1.Metode Min-Max Scaling

```python
from sklearn.preprocessing import MinMaxScaler
df_target = df.copy()

scaler = MinMaxScaler(feature_range=(0,1))
df_target[colom_normals] = scaler.fit_transform(df_target[colom_normals])

print(f'Sesudah Minmax Scaler:\n {df_target[colom_normals].head(5)}')
```

### 2.Metode MaxAbs Scaling

```python
from sklearn.preprocessing import MaxAbsScaler
df_target = df.copy()

scaler = MaxAbsScaler()
df_target[colom_normals] = scaler.fit_transform(df_target[colom_normals])

print(f'Sesudah MaxAbs Scaler:\n {df_target[colom_normals].head(5)}')
```

### 3.Metode Robust Scaling

```python
from sklearn.preprocessing import RobustScaler
df_target = df.copy()

scaler = RobustScaler()
df_target[colom_numerik] = scaler.fit_transform(df_target[colom_numerik])

print(f'Sesudah Robust Scaling:\n {df_target[colom_numerik].head(5)}')
```

### 4.Metode Vector Transformation

```python
from sklearn.preprocessing import normalize
df_target = df.copy()

df_target = df_target.dropna()
df_target[colom_numerik] = normalize(df_target[colom_numerik],norm='l2') # bisa pilih l1 or l2

print(f'Sesudah Vector Trasnfrom :\n {df_target[colom_numerik].head(5)}')
```

## 2.Standardisasi Data

### 1.Metode Z-Score Standardization 

```python
from sklearn.preprocessing import StandardScaler
df_target = df.copy()

scaler = StandardScaler()
df_target[colom_normals] = scaler.fit_transform(df_target[colom_normals])
print(f'Sesudah Z-score Standardization :\n {df_target[colom_normals].head(5)}')
```

### 2.Metode Mean Normalization

```python
df_target = df.copy()
df_target[colom_normals] = (df_target[colom_normals]-df_target[colom_normals].mean()) / (df_target[colom_normals].max() - df_target[colom_normals].min())
df_target[colom_normals]
```

### 3.Unit Variance Scaling

```python
from sklearn.preprocessing import StandardScaler
df_target = df.copy()

scaler = StandardScaler(with_mean=True,with_std=True)
df_target[colom_normals] = scaler.fit_transform(df_target[colom_normals])
print(f'Sesudah Z-score Standardization :\n {df_target[colom_normals].head(5)}')
```

## 3.Transformasi Yeo-Johnson

```python
skewsness = df.select_dtypes(include=np.number).skew()
colom_skew = skewsness[skewsness.abs() >= 0.5].index
```

```python
from sklearn.preprocessing import PowerTransformer
df_target = df.copy()

pt = PowerTransformer(method='yeo-johnson')
df_target[colom_skew] = pt.fit_transform(df_target[colom_skew])

print(f'Skewsness Sebelum di Transfrom: {df[colom_skew].skew().iloc[0].round(2)}')
print(f'Skewsness Sesudah di Transfrom: {df_target[colom_skew].skew().iloc[0].round(2)}')
```
