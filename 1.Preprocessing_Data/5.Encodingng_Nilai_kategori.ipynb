```python
import pandas as pd
```

```python
df = pd.read_csv('../0.dataset/dataset_karyawan_1000.csv')
df.head(10)
```

```python
colom_kategori = df.select_dtypes(include='object').drop(columns=['ID_Karyawan']).columns
```

## 1.One-Hot Encoding

### 1.Simple One-Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
df_target = df.copy()

encoder = OneHotEncoder(sparse_output=False,drop=None)
encoded_features = encoder.fit_transform(df_target[colom_kategori])
encoded_df = pd.DataFrame(encoded_features,columns=encoder.get_feature_names_out(colom_kategori))

encoded_df.head(5)
```

### 2.One-Hot Encoding dengan Drop First

```python
df_target = df.copy()

df_target = pd.get_dummies(df_target,columns=colom_kategori,dtype=int)
df_target.head()
```

## 2.Label Encoding

### 1.Label Encoding Sederhana

```python
from sklearn.preprocessing import LabelEncoder
df_target = df.copy()

label = LabelEncoder()

for col in colom_kategori:
    df_target[col] = label.fit_transform(df_target[col])

df_target[colom_kategori].head()
```

### 2.Label Encoding Berdasarkan Frekuensi

```python
df_target = df.copy()

for col in colom_kategori:
    frequnecy = df_target[col].value_counts()
    df_target[col] = df_target[col].map(frequnecy)

df_target[colom_kategori].head()
```

## 3.Binary Encoding

### 1.Binary Encoding Sederhana

```python
from category_encoders import BinaryEncoder
df_target = df.copy()

encoder = BinaryEncoder(cols=colom_kategori)
encoded_data = encoder.fit_transform(df_target[colom_kategori])

encoded_data.head(5)
```

## 4.Ordinal Encoding

### 1.Ordinal Encoding Berdasarkan Urutan Logis

```python
from sklearn.preprocessing import OrdinalEncoder
colom_ordinal = ['Pendidikan_Terakhir','Level_Jabatan']
df_target = df[colom_ordinal].copy()

urutan_akademik = ['SMA/SMK','D3','S1','S2']
urutan_jabatan = ['Junior','Mid-Level','Senior','Lead/Manager']

df_target['Level_Jabatan'].unique()
encoder = OrdinalEncoder(categories=[urutan_akademik,urutan_jabatan],dtype=int)

df_target[colom_ordinal] = encoder.fit_transform(df_target[colom_ordinal])

df_target.head()
```

## 5.Frequency Encoding

### 1.Frequency Encoding Sederhana

```python
df_target = df.copy()

for col in colom_kategori:
    frequnecy = df_target[col].value_counts()
    df_target[col] = df_target[col].map(frequnecy)

df_target[colom_kategori].head()
```

### 2.Frequency Encoding dengan Normalisasi

```python
df_target = df.copy()
from sklearn.preprocessing import MinMaxScaler

for col in colom_kategori:
    frequnecy = df_target[col].value_counts()
    df_target[col] = df_target[col].map(frequnecy)

scaler = MinMaxScaler()
df_target[colom_kategori]= scaler.fit_transform(df_target[colom_kategori]).round(2)

df_target[colom_kategori].head(5)
```

## 6.Hash Encoding

### 1.Hash Encoding Sederhana

```python
df_target = df.copy()
from category_encoders import HashingEncoder

encoder = HashingEncoder(cols=colom_kategori,n_components=3)
df_target = encoder.fit_transform(df_target)

df_target.head(5)
```

### 2.Hash Encoding dengan Collision Handling

```python
df_target = df.copy()
from category_encoders import HashingEncoder

encoder = HashingEncoder(cols=colom_kategori,n_components=16)
df_target = encoder.fit_transform(df_target)

df_target.head()
```
