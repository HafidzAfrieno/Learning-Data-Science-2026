### 9. Agregasi dan Pengelompokan dengan Pandas

```python
data = {'Category': ['A', 'B', 'A', 'B'], 'Value': [10, 20, 30, 40]}
df = pd.DataFrame(data)

grouped_sum = df.groupby('Category').sum()
print(f"Sum:\n {grouped_sum}")

grouped_mean = df.groupby('Category').mean()
print(f"Mean:\n {grouped_mean}")

grouped_agg = df.groupby('Category').agg(['sum','mean'])
print(f"Aggregated:\n {grouped_agg}")
```

### 10. Menggabungkan dan Menyusun Data di Pandas

```python
# 1. Menggabungkan DataFrame
data1 = {
    'Name': ['Alice', 'Charlie', 'Edward', 'Grace'],
    'Years_Experience': [2, 3, 4, 6]
}
df1 = pd.DataFrame(data1)

data2 = {
    'Name': ['Alice', 'Charlie', 'Frank', 'Grace'],
    'Role': ['Manager', 'Analyst', 'Developer', 'HR']
}
df3 = pd.DataFrame(data2)

df_merged_inner = pd.merge(df1,df3,on='Name',how='inner')
df_merged_outer = pd.merge(df1,df3,on='Name',how='outer')
df_merged_left = pd.merge(df1,df3,on='Name',how='left')
df_merged_right = pd.merge(df1,df3,on='Name',how='right')

print(f"\nInner merge (Intersection of Names):\n {df_merged_inner}")
print(f"\nOuter Merge (Union of Names):\n {df_merged_outer}")
print(f"\nLeft Merge (Left of Names):\n {df_merged_left}")
print(f"\nRight Merge (Right of Names):\n {df_merged_right}")
```

```python
# 2. Menggabungkan Data vertikal
df = pd.DataFrame({
    'Name': ['Alice', 'Charlie', 'Edward', 'Grace'],
    'Years_Experience': [2, 3, 4, 6],
    'Role': ['Manager', 'Analyst', 'Developer', 'HR']
})
new_df = pd.DataFrame({
    'Name': ['John', 'Lily'],
    'Years_Experience': [5, 3],
    'Role': ['Designer', 'Developer']
})
concated_df = pd.concat([df,new_df],ignore_index=True)

print("Conatenated Dataframe:\n")
print(concated_df)
```

### 11. Membentuk Ulang Data di Pandas

```python
# 1. Tabel Pivot
data = {'Date': ['2024-10-03', '2024-10-01', '2024-10-02', '2024-10-02'],'Category': ['A', 'B', 'A', 'B'],'Values': [10, 20, 15, 25]}
df = pd.DataFrame(data)

pivot_table = df.pivot_table(values='Values',index='Date',columns='Category')
print(pivot_table)
```

```python
# 2. Penggabungan (Melting)
data = {
    'Date': ['2024-10-01', '2024-10-01', '2024-10-02', '2024-10-02'],
    'Category_A': [10, None, 15, None],
    'Category_B': [None, 20, None, 25]
}
df = pd.DataFrame(data)

melt_df = df.melt(id_vars='Date',var_name='Kategori',value_name='Nilai')
melt_df
```

```python
# 3. Menumpuk dan Membongkar Tumpukan dengan Pandas
data = {
    ('Sales', 'Q1'): [100, 150],
    ('Sales', 'Q2'): [200, 250],
    ('Expenses', 'Q1'): [50, 70],
    ('Expenses', 'Q2'): [80, 90]
}
index = ['Product_A','Product_B']
df = pd.DataFrame(data,index=index)
print(df)

staked_df = df.stack()
print(f"\nStaked:\n {staked_df}")

unstaked_df = staked_df.unstack()
print(f"\nUnstaked:\n {unstaked_df}")
```

```python
# meneukan penggunaan memori
df = pd.read_csv('../0.Dataset/data_salary_rice_50000.csv')
df.info(memory_usage='deep')
```

```python
#melihat jumlah pemakaian memori
df.memory_usage()
```

```python
#Mengubah kolom numerik ke tipe data yang lebih kecil:
data = df.copy()
numerik_colom = ['salary_rice_kg','allowance_value_idr']

sebelum = data[numerik_colom].memory_usage()

data[numerik_colom]=data[numerik_colom].astype('int16')
sesudah = data[numerik_colom].memory_usage()

print(f"Sebelum:\n {sebelum}")
print(f"Sesudah:\n {sesudah}")
```
