```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
data = load_iris()
X = pd.DataFrame(data = data.data,columns=data.feature_names)
y = data.target

#splt data into trainign and testing sets
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

## Pemilihan Fitur dengan Scikit-Learn

### 1. Seleksi Univariat di Python dengan Scikit-Learn

```python
from sklearn.feature_selection import SelectKBest,chi2

select_k_best = SelectKBest(score_func=chi2,k=3)
X_train_k_best = select_k_best.fit_transform(X_train,y_train)

print(f"Selected Features:\n {X_train.columns[select_k_best.get_support()]}")
```

### 2. Eliminasi Fitur Rekursif

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
rfe = RFE(model,n_features_to_select=3)
X_train_rfe = rfe.fit_transform(X_train,y_train)

print(f"Selected Features:\n {X_train.columns[rfe.get_support()]}")
```

### 3. Penentuan Pentingnya Fitur Berdasarkan Pohon

```python
from sklearn.ensemble import RandomForestClassifier

# Train random forest and get feature importances
model = RandomForestClassifier()
model.fit(X_train,y_train)
importances = model.feature_importances_

feature_importances = pd.Series(importances,index=X_train.columns)
print(feature_importances.sort_values(ascending=False))
```
