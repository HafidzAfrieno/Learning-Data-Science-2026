```python
from sklearn.preprocessing import OrdinalEncoder
import pandas as pd
```

### 1. Melakukan Encoding pada data ordinal

```python
data = pd.DataFrame({ 
    'Satisfaction': ['Low', 'High', 'Medium', 'Low', 'High'], 
    'Score': [3, 8, 5, 2, 9] 
})

categories_order = [['Low','Medium','High']]

encoder = OrdinalEncoder(categories=categories_order)
data['Satisfaction_Encoded'] = encoder.fit_transform(data[['Satisfaction']])
print(data)
```
