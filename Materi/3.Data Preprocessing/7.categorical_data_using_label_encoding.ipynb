```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

## LabelEncoder di Scikit-Learn

### 1. Pengkodean Label Numerik

```python

label = LabelEncoder()
label.fit([1,2,2,4,4,5,6,7,7])

input = [1,4,5,2,2,6,1,1,6,7,5]
output = label.transform(input)

print(f"Nilai Uniqe: {label.classes_}")
print(f"Convert Nilai {input} jadi: {output}")
print(f"Inverse {output} kenilai aslinya: {label.inverse_transform(output)}")
```

### 2. Pengkodean Label String

```python
label = LabelEncoder()
label.fit(["paris", "paris", "tokyo", "amsterdam"])

input = ["tokyo", "tokyo", "paris","amsterdam","paris"]
output = label.transform(input)

print(f"Nilai Uniqe: {label.classes_}")
print(f"Convert Nilai {input} jadi: {output}")
print(f"Inverse {output} kenilai aslinya: {label.inverse_transform(output)}")
```

### 3. Menggunakan Label Encoder pada Data Frame

```python
data = pd.DataFrame({
    'Fruit': ['Apple', 'Banana', 'Orange', 'Apple', 'Orange', 'Banana'],
    'Price': [1.2, 0.5, 0.8, 1.3, 0.9, 0.6]
})
label = LabelEncoder()
data['Fruit_encoded'] = label.fit_transform(data['Fruit'])
print(data)
```

### 4. Menangani Kategori yang Tidak Terlihat dalam Data Uji
menerapkan nilai default -1 untuk data ordinal yang tidak ada pada data train

```python
train = pd.DataFrame({'City': ['Delhi', 'Mumbai', 'Chennai', 'Delhi']})
test = pd.DataFrame({'City': ['Mumbai', 'Kolkata','jakarta']})

label = LabelEncoder()
train['Cith_encoded']= label.fit_transform(train['City'])
test['City_encoded'] = test['City'].apply(lambda x: label.transform([x])[0] if x in label.classes_ else -1)

print(train)
print(test)
```
