```python
import pandas as pd
```

### 1. Pandas dataframe.drop_duplicates()

```python
data = {
    "Name": ["Alice", "Bob", "Alice", "David"],
    "Age": [25, 30, 25, 40],
    "City": ["NY", "LA", "NY", "Chicago"]
}

df = pd.DataFrame(data)
print(f"Original Dataframe:\n {df}")

df_cleaned = df.drop_duplicates()
print(f"\nModified Dataframe (no duplicated):\n {df_cleaned}")
```

### 2. Menghapus Data Duplikat Berdasarkan Kolom Tertentu


```python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Alice', 'David'],
    'Age': [25, 30, 25, 40],
    'City': ['NY', 'LA', 'SF', 'Chicago']
})
df_cleaned = df.drop_duplicates(subset=['Name'])
print(df_cleaned)
```

### 3. Mempertahankan Kemunculan Terakhir dari Duplikat

```python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Alice', 'David'],
    'Age': [25, 30, 25, 40],
    'City': ['NY', 'LA', 'SF', 'Chicago']
})
df_cleaned = df.drop_duplicates(subset=['Name'],keep='last')
print(df_cleaned)
```

### 3. Menghapus Semua Duplikat

```python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Alice', 'David'],
    'Age': [25, 30, 25, 40],
    'City': ['NY', 'LA', 'NY', 'Chicago']
})
df_cleaned = df.drop_duplicates(keep=False)
print(df_cleaned)
```

### 4. Menghapus Duplikat Berdasarkan Kolom yang Sebagian Identik

```python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Alice', 'David'],
    'Age': [25, 30, 25, 40],
    'City': ['NY', 'LA', 'NY', 'Chicago']
})
df_cleaned = df.drop_duplicates(subset=['Name','City'])
print(df_cleaned)
```
