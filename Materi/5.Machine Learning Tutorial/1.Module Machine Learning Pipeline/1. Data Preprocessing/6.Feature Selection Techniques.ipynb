# Panduan Lengkap Feature Selection dalam Machine Learning

*Feature selection* (pemilihan fitur) adalah proses krusial untuk memilih variabel atau input yang paling relevan untuk digunakan dalam model *machine learning*. Langkah ini membantu mengurangi noise, mempercepat waktu training, dan menghindari *overfitting*.

---

## 3 Metode Utama Feature Selection

### 🔍 1. Filter Methods

* **Penjelasan:** Metode ini menyeleksi fitur di tahap awal (*preprocessing*) sebelum model *machine learning* dilatih. Caranya dengan melihat karakteristik statistik data (seperti korelasi) secara independen terhadap variabel target. Karena tidak bergantung pada algoritma model, metode ini sangat cepat dan efisien.
* **Analogi/Contoh:** Seperti menyaring pasir menggunakan ayakan sebelum membuat semen; batu-batu besar langsung dibuang dari awal tanpa perlu tahu struktur bangunan apa yang akan dibuat.

> 📌 **Catatan Penting:** Karena memeriksa fitur secara mandiri, *Filter Methods* terkadang bisa melewatkan interaksi berharga antar kombinasi fitur yang sebenarnya berguna jika digabungkan.

Beberapa teknik populer dalam *Filter Methods*:

* [Information Gain](https://www.geeksforgeeks.org/machine-learning/feature-selection-techniques-in-machine-learning/): Mengukur seberapa banyak informasi yang diberikan fitur untuk memprediksi target.
* [Chi-Square Test](https://www.geeksforgeeks.org/machine-learning/feature-selection-techniques-in-machine-learning/): Digunakan untuk melihat hubungan antara dua fitur kategorikal.
* [Pearson’s Correlation Coefficient](https://www.geeksforgeeks.org/machine-learning/feature-selection-techniques-in-machine-learning/): Mengukur kekuatan hubungan linier antara dua variabel kontinu.

---

### 🔄 2. Wrapper Methods

* **Penjelasan:** Metode ini memperlakukan pemilihan fitur sebagai masalah pencarian (*search problem*). Ia akan mencoba berbagai kombinasi subset fitur, melatih model dengan subset tersebut, lalu mengevaluasi hasilnya. Proses penambahan atau pengurangan fitur akan terus dilakukan secara berulang (rakus/*greedy*) hingga performa terbaik tercapai.
* **Analogi/Contoh:** Seperti mencoba berbagai kombinasi baju, celana, dan sepatu di ruang ganti untuk menemukan perpaduan pakaian yang paling cocok untuk acara malam ini.

Beberapa teknik populer dalam *Wrapper Methods*:

* **Forward Selection:** Dimulai dari nol fitur, lalu menambahkan satu per satu fitur yang paling meningkatkan performa model.
* **Backward Elimination:** Dimulai dengan memasukkan semua fitur, lalu membuang satu per satu fitur yang paling tidak berguna.
* [Recursive Feature Elimination (RFE)](https://www.geeksforgeeks.org/machine-learning/feature-selection-techniques-in-machine-learning/): Membuang fitur yang memiliki bobot paling rendah secara bertahap hingga menyisakan jumlah fitur yang diinginkan.

---

### 🧠 3. Embedded Methods

* **Penjelasan:** Metode ini melakukan pemilihan fitur secara otomatis **bersamaan** dengan proses training model berlangsung. Algoritma bawaan model ini memiliki mekanisme internal sendiri untuk menentukan fitur mana yang paling berkontribusi dalam mengurangi error prediksi.
* **Analogi/Contoh:** Seperti seorang koki profesional yang secara otomatis menyesuaikan takaran bumbu dan membuang bahan yang dirasa tidak perlu langsung saat dia sedang memasak di atas wajan.

Beberapa teknik populer dalam *Embedded Methods*:

* **L1 Regularization (Lasso):** Menambahkan penalti yang bisa memaksa koefisien fitur yang tidak penting menjadi benar-benar nol (dieliminasi).
* **Decision Trees & Random Forests:** Mengukur tingkat pentingnya fitur berdasarkan seberapa banyak fitur tersebut membantu memurnikan (*impurity reduction*) percabangan data.
