### 2. Membaca File JSON Bersarang

```python
# Menggunakan Modul json dan metode pd.json_normalize()
import json
data = {"One": {"0": 60, "1": 60, "2": 60, "3": 45, "4": 45, "5": 60},
        "Two": {"0": 110, "1": 117, "2": 103, "3": 109, "4": 117, "5": 102}}

json_data = json.dumps(data)

df_normalize = pd.json_normalize(json.loads(json_data))
print(f"\nDataframe using Json Module and 'ps.json_normalize()'\n {df_normalize}")

df = pd.DataFrame(df_normalize)
df
```
