```python
import pandas as pd
import numpy as np
```

## Memeriksa Nilai yang Hilang di Pandas

### 1. Menggunakan isnull()

```python
# 1. Menemukan Nilai yang Hilang dalam DataFrame
data = {'First Score': [100, 90, np.nan, 95],
        'Second Score': [30, 45, 56, np.nan],
        'Third Score': [np.nan, 40, 80, 98]}
df = pd.DataFrame(data)
mv = df.isnull()
print(mv)
```

```python
# 2.  Menemukan Nilai yang Hilang dalam DataFrame
data = pd.read_csv('../0.Dataset/employees.csv')
bool_series = pd.isnull(data['First Name'])
missing_gender_data = data[bool_series]
print(missing_gender_data)
```

### 2. Menggunakan isna()

```python
data = {'Name': ['Amit', 'Sita', np.nan, 'Raj'],
        'Age': [25, np.nan, 22, 28]}

df = pd.DataFrame(data)

print(df.isna())
```

### 3. Memeriksa Nilai yang Tidak Hilang Menggunakan notnull()

```python
# 1. Mengidentifikasi Nilai yang Tidak Hilang dalam DataFrame
ata = {'First Score': [100, 90, np.nan, 95],
        'Second Score': [30, 45, 56, np.nan],
        'Third Score': [np.nan, 40, 80, 98]}

df = pd.DataFrame(data)
Non_missingValue = df.notnull()

print(Non_missingValue)
```

```python
# 2. Memfilter Data dengan Nilai yang Tidak Hilang
data = pd.read_csv('../0.Dataset/employees.csv')
notnull = pd.notnull(data['Gender'])

gender_notnull = data.loc[notnull]
print(gender_notnull)
```

## Mengisi Nilai yang Hilang di Pandas

### 1. Menggunakan fillna()

```python
# 1. isi Nilai yang Hilang dengan Nol
data = {'First Score': [100, 90, np.nan, 95],
        'Second Score': [30, 45, 56, np.nan],
        'Third Score': [np.nan, 40, 80, 98]}

df = pd.DataFrame(data)
df= df.fillna(0)
print(df)
```

```python
# 2. Isi dengan Nilai Sebelumnya (Isi Maju)
data = {'First Score': [100, 90, np.nan, 95],
        'Second Score': [30, 45, 56, np.nan],
        'Third Score': [np.nan, 40, 80, 98]}

df = pd.DataFrame(data)
df= df.ffill()
print(df)
```

```python
# 3. Isi dengan Nilai Berikutnya (Isi Mundur)
data = {'First Score': [100, 90, np.nan, 95],
        'Second Score': [30, 45, 56, np.nan],
        'Third Score': [np.nan, 40, 80, 98]}

df = pd.DataFrame(data)
df= df.bfill()
print(df)
```

```python
# 4. Isi Nilai NaN dengan 'Tidak Ada Jenis Kelamin'
data = pd.read_csv('../0.Dataset/employees.csv')
data= data[10:25]

data['Gender'] = data['Gender'].fillna('No Gender')
data
```

### 2. Menggunakan replace()
Gunakan fungsi replace() untuk mengganti nilai NaN dengan nilai tertentu.

```python
data = pd.read_csv('../0.Dataset/employees.csv')

data = data.replace(to_replace=np.nan,value=-999)
data[10:25]
```

### 3. Menggunakan interpolate()
Fungsi interpolate() mengisi nilai yang hilang menggunakan teknik interpolasi seperti metode linier.

```python
df = pd.DataFrame({"A": [12, 4, 5, None, 1], 
                   "B": [None, 2, 54, 3, None], 
                   "C": [20, 16, None, 3, 8], 
                   "D": [14, 3, None, None, 6]})  
print(f"Sebelum:\n {df}")
df = df.interpolate(method='linear',limit_direction='forward')
print(f"Sesudah:\n {df}")
```
