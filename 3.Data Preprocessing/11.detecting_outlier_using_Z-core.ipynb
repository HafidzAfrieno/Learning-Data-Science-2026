```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import zscore
```

```python
data = [5, 2, 4.5, 4, 3, 2, 6, 20, 9, 2.5, 3.5, 4.75, 6.5, 2.5, 8, 1,40,2,5,36,2,4,7]
df = pd.DataFrame(data, columns=['Value'])
```

```python
df['Z-score'] = zscore(df['Value'])
df
```

```python
outliers = df.loc[df['Z-score'].abs() > 3 ] #3 adalah batas
outliers
```

```python
plt.figure(figsize=(10,6))

plt.scatter(df['Value'].index,df['Z-score'],label='Data Points')
plt.scatter(outliers['Value'].index,outliers['Z-score'],color='red',label='Outliers')

plt.xlabel('Index Value')
plt.ylabel('Z-score')
plt.title('Scatter Plot of Value vs. Z-score')
plt.legend()
plt.grid(True)
plt.show()
```
