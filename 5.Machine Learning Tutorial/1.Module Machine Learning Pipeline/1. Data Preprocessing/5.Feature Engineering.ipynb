# Catatan Kuliah: Feature Engineering dalam Data Science

> 📌 **Catatan Penting:** Feature engineering adalah kunci utama yang menentukan performa model machine learning. Model secanggih apa pun tidak akan memberikan hasil maksimal jika data yang dimasukkan tidak representatif.

---

## 🛠️ Apa itu Feature Engineering?

### 🧩 Definisi Dasar

* **Penjelasan:** Proses memilih, mengubah, atau membuat variabel baru (fitur) dari data mentah agar model machine learning bisa mempelajari pola data dengan jauh lebih mudah dan akurat.
* **Analogi/Contoh:** Seperti seorang koki yang mengupas, memotong, dan membumbui bahan makanan mentah sebelum dimasak agar hidangan akhirnya terasa lezat.

### 🌟 Mengapa Ini Penting?

* **Penjelasan:** Langkah ini meningkatkan akurasi model, mencegah *overfitting* (model terlalu menghafal data latihan), membuat model lebih cepat saat dilatih, dan membuat hasil prediksi lebih mudah dipahami oleh manusia.
* **Analogi/Contoh:** Memilih baju yang pas dan sesuai ukuran tubuh membuat kita bergerak lebih lincah dan terlihat rapi, sama seperti memilih fitur penting yang membuat model bekerja efisien.

---

## 🔄 Proses Utama dalam Feature Engineering

### ⚡ 1. Feature Creation (Membuat Fitur Baru)

* **Penjelasan:** Membuat variabel baru dari data yang sudah ada, baik berdasarkan pengetahuan bisnis (domain) maupun dari kombinasi data yang tersedia.
* **Analogi/Contoh:** Menggabungkan data `Panjang` dan `Lebar` tanah untuk membuat fitur baru bernama `Luas_Tanah`.

### 🔄 2. Feature Transformation (Transformasi Fitur)

* **Penjelasan:** Mengubah bentuk atau skala nilai data asli agar lebih ramah terhadap algoritma machine learning (misalnya mengubah teks menjadi angka atau meratakan data yang melenceng jauh).
* **Analogi/Contoh:** Mengubah ukuran mata uang dari Rupiah ke Dollar agar angkanya tidak terlalu besar saat dihitung model.

### ✂️ 3. Feature Extraction (Ekstraksi Fitur)

* **Penjelasan:** Memadatkan banyak fitur lama menjadi beberapa fitur baru yang lebih ringkas tanpa kehilangan informasi penting (misalnya menggunakan metode PCA).
* **Analogi/Contoh:** Merangkum satu bab buku tebal menjadi satu halaman catatan kecil yang padat inti sarinya.

### 🎯 4. Feature Selection (Pemilihan Fitur)

* **Penjelasan:** Menyaring dan hanya memilih fitur yang paling relevan serta membuang fitur yang tidak penting atau bikin bingung model.
* **Analogi/Contoh:** Saat packing baju untuk liburan 3 hari, kamu hanya membawa baju yang pasti dipakai dan meninggalkan sisanya di rumah.

### 📏 5. Feature Scaling (Standardisasi Nilai)

* **Penjelasan:** Menyamakan rentang nilai antar-fitur agar adil dan tidak ada satu fitur yang mendominasi fitur lainnya hanya karena angka dasarnya besar.
* **Analogi/Contoh:** Mengubah nilai ujian berskala 0-100 dan skala 0-4 (IPK) ke dalam format persentase agar nilainya setara dan bisa dibandingkan.

## 💻 Teknik Populer & Implementasi Kode Python

Berikut adalah beberapa teknik umum beserta contoh kode [Pandas](https://www.geeksforgeeks.org/machine-learning/what-is-feature-engineering/) yang siap kamu jalankan di Jupyter Notebook:

### 📊 One-Hot Encoding

* **Penjelasan:** Mengubah data kategori teks menjadi angka biner (0 dan 1) agar bisa dibaca oleh model matematika.

```python
import pandas as pd

# Membuat data tiruan berbentuk teks
data = {'Color': ['Red', 'Blue', 'Green', 'Blue']}
df = pd.DataFrame(data)

# Mengubah teks menjadi kolom biner 0 dan 1
df_encoded = pd.get_dummies(df, columns=['Color'], prefix='Color')
print(df_encoded)

```

* **Penjelasan Kode:** * `pd.get_dummies(...)` digunakan untuk memecah kolom teks `Color` menjadi 3 kolom baru (`Color_Red`, `Color_Blue`, `Color_Green`).
* Jika baris data aslinya berwarna 'Red', maka kolom `Color_Red` akan bernilai `True` (atau 1) dan sisanya `False` (atau 0).


* **Analogi/Contoh:** Seperti lembar absensi kelas, jika kamu hadir maka namamu dicentang (1), jika tidak hadir diberi tanda kosong (0).

### 🗑️ Binning (Pengelompokan Nilai)

* **Penjelasan:** Mengelompokkan data angka yang kontinu (terus-menerus) ke dalam beberapa kategori atau "kotak" tertentu.

```python
import pandas as pd

# Data umur pelanggan
data = {'Age': [23, 45, 18, 34, 67, 50, 21]}
df = pd.DataFrame(data)

# Menentukan batas kotak dan labelnya
bins = [0, 20, 40, 60, 100]
labels = ['Remaja', 'Dewasa_Muda', 'Paruh_Baya', 'Lansia']

# Memasukkan umur ke dalam kelompok yang sesuai
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
print(df)

```

* **Penjelasan Kode:** * `bins` menentukan batas umur (misal: 0 sampai sebelum 20).
* `pd.cut(...)` bertugas memotong angka umur tersebut dan melabelinya otomatis menjadi kelompok seperti 'Remaja' atau 'Lansia'.


* **Analogi/Contoh:** Memisahkan telur ayam ke dalam wadah ukuran "Kecil", "Sedang", dan "Besar" berdasarkan beratnya.

## 📐 Penjelasan Rumus Scaling Secara Sederhana

Dalam melakukan *Feature Scaling*, ada dua rumus yang sangat sering digunakan:

### 1. Min-Max Scaling (Normalisasi)

Mengubah data agar berada tepat di dalam rentang 0 sampai 1.

$$X_{new} = \frac{X - X_{min}}{X_{max} - X_{min}}$$

* **Cara Gampang Membaca Rumus:** > "Nilai baru didapat dari: nilai asli dikurangi nilai paling kecil, lalu dibagi dengan selisih antara nilai paling besar dan paling kecil di kumpulan data tersebut."

### 2. Standard Scaling (Standardisasi)

Mengubah data agar berpusat di angka 0 dengan penyebaran (varians) bernilai 1.

$$X_{new} = \frac{X - \mu}{\sigma}$$

* **Cara Gampang Membaca Rumus:** > "Nilai baru didapat dari: nilai asli dikurangi rata-rata ($\mu$), lalu dibagi dengan simpangan baku ($\sigma$) atau ukuran standar penyebaran datanya."
