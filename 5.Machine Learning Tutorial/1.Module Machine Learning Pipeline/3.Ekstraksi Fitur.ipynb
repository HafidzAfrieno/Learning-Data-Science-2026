# Ekstraksi Fitur (Feature Extraction) dalam Machine Learning

Ekstraksi fitur adalah proses mengubah data mentah yang kompleks dan tidak terstruktur menjadi sekumpulan variasi data baru yang lebih ringkas, bermakna, dan terstruktur. Tujuannya adalah agar model machine learning dapat mengenali pola dan hubungan antar-data dengan jauh lebih mudah dan cepat.

---

## 💡 Mengapa Ekstraksi Fitur itu Penting?

* **Mengurangi Beban Komputasi:** Menyederhanakan data mentah yang rumit sehingga komputer tidak perlu bekerja terlalu keras.
* **Meningkatkan Performa Model:** Hanya fokus pada data yang relevan, sehingga akurasi prediksi meningkat.
* **Menghindari Overfitting:** Membuang "noise" atau gangguan tidak penting agar model tidak terlalu menghafal data latihan.

> ⚠️ **Catatan Penting:** Ekstraksi Fitur **berbeda** dengan Pemilihan Fitur (Feature Selection).
> * **Feature Selection:** Memilih sebagian fitur asli yang dianggap penting tanpa mengubah bentuknya.
> * **Feature Extraction:** Menggabungkan atau mengubah fitur asli menjadi fitur baru yang benar-benar berbeda melalui perhitungan tertentu.

## 🛠️ Teknik-Teknik Utama Ekstraksi Fitur

### 📊 1. Metode Statistik

* **Penjelasan:** Menggunakan nilai-nilai statistik dasar seperti Rata-rata (Mean), Median, Deviasi Standar, serta Korelasi untuk meringkas tren utama dan sebaran dari sekumpulan data.
* **Analogi/Contoh:** Seperti melihat nilai rata-rata rapor seorang siswa untuk menilai kemampuan akademiknya secara umum, ketimbang melihat nilai tugasnya satu per satu tiap minggu.

### 📉 2. Pengurangan Dimensi (Dimensionality Reduction)

* **Penjelasan:** Memperkecil jumlah variabel atau kolom data tanpa kehilangan informasi-informasi penting di dalamnya. Teknik yang populer digunakan antara lain [Analisis Komponen Utama (PCA)](https://www.geeksforgeeks.org/machine-learning/what-is-feature-extraction/), LDA, dan t-SNE.
* **Analogi/Contoh:** Seperti memotret benda 3 dimensi menjadi foto 2 dimensi; objeknya menjadi lebih sederhana untuk dilihat tetapi bentuk aslinya tetap bisa kita kenali dengan jelas.

### 🔤 3. Ekstraksi untuk Data Tekstual (NLP)

* **Penjelasan:** Mengubah teks atau kalimat mentah menjadi deretan angka agar bisa dibaca komputer. Contohnya menggunakan *Bag of Words (BoW)* (menghitung frekuensi kata) atau *TF-IDF* (pembobotan kata unik).
* **Analogi/Contoh:** Sistem penyaring *spam* email yang mendeteksi kata "Diskon" atau "Hadiah" yang muncul berulang kali untuk menandai bahwa email tersebut adalah spam.

### 🎵 4. Pemrosesan Sinyal (Signal Processing)

* **Penjelasan:** Mengonversi data berbasis waktu (seperti audio atau sensor) menggunakan metode seperti *Transformasi Fourier* atau *Transformasi Wavelet* untuk menganalisis komponen frekuensi di dalamnya.
* **Analogi/Contoh:** Aplikasi Shazam yang mendengarkan potongan lagu di ruangan bising, lalu mengekstrak frekuensi unik suara lagu tersebut untuk mencocokkannya dengan judul lagu asli.

### 📷 5. Ekstraksi Data Gambar (Computer Vision)

* **Penjelasan:** Mengambil karakteristik visual penting dari sebuah gambar, misalnya menggunakan algoritma *HOG (Histogram of Oriented Gradients)* untuk mendeteksi tepian objek, atau menggunakan lapisan konvolusi pada *CNN (Convolutional Neural Network)*.
* **Analogi/Contoh:** Sistem mengemudi otomatis pada mobil pintar yang mengekstrak bentuk lingkaran dan warna merah dari kamera untuk mengenali bahwa di depannya ada rambu "Stop".

## 🧠 Rumus & Cara Membaca Mudahnya

Salah satu metode yang sering digunakan dalam pembobotan teks adalah **TF-IDF**. Mari kita lihat rumusnya dengan cara yang paling santai:

$$\text{Skor TF-IDF} = \text{TF} \times \text{IDF}$$

> 💡 **Cara Gampang Membaca Rumus:**
> "Bobot pentingnya sebuah kata itu didapat dari **seberapa sering kata itu muncul di satu dokumen (TF)** dikalikan dengan **seberapa langka kata tersebut di dokumen-dokumen lainnya (IDF)**. Semakin sering muncul di satu tempat tapi langka di tempat lain, artinya kata itu sangat spesial!"

## 🗺️ Alur Kerja Ekstraksi Fitur (Kesimpulan)

Secara matematis dan logika pemrograman, alur kerja transformasi data mentah menjadi bentuk fitur siap pakai dapat disimpulkan dalam fungsi komposisi sederhana berikut:

$$\text{Data Mentah} \longrightarrow f(\text{Transformasi}) \longrightarrow \text{Fitur Baru}$$

Jika dituliskan dalam alur kerja sistemik di Jupyter Notebook:

$$\mathbf{X}_{\text{mentah}} \xrightarrow{\quad \text{Pre-processing} \quad} \mathbf{X}_{\text{bersih}} \xrightarrow{\quad \phi(\mathbf{X}) \quad} \mathbf{X}_{\text{fitur}} \xrightarrow{\quad \text{Model ML} \quad} \mathbf{Y}_{\text{prediksi}}$$
