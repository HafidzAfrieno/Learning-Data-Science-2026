## 5 Teknik Utama Feature Scaling

### 📈 1. Absolute Maximum Scaling

* **Penjelasan:** Teknik ini membagi setiap nilai dengan nilai absolut terbesar dari fitur tersebut. Hasil akhirnya akan selalu berada dalam rentang $-1$ hingga $1$. Sangat cocok digunakan untuk data yang jarang (*sparse data*) tapi sangat sensitif terhadap pencilan (*outliers*).
* **Analogi/Contoh:** Seperti membandingkan tinggi badan semua orang di kelas dengan tinggi pemain basket tertinggi di kelas tersebut sebagai acuan utama.
* **Cara Membaca Rumus:**

$$X_{scaled} = \frac{X_i}{\max(|X|)}$$



*Cara gampang membacanya:* "Nilai baru adalah nilai asli saat ini dibagi dengan angka mutlak paling besar yang ada di kolom tersebut."
* **Penjelasan Kode Python:**
```python
max_abs = np.max(np.abs(df), axis=0) # Mencari nilai absolut terbesar di setiap kolom
scaled_df = df / max_abs             # Membagi seluruh isi data dengan nilai terbesar tersebut

```



### 🎯 2. Min-Max Scaling (Normalization)

* **Penjelasan:** Mengubah dan menyusutkan data agar berada dalam rentang yang pasti, biasanya antara $0$ hingga $1$. Distribusi asli data tetap terjaga, namun teknik ini sangat rapuh jika data Anda memiliki *outliers* yang ekstrem.
* **Analogi/Contoh:** Mengubah nilai ujian dari skala 0-100 menjadi skala IPK 0-4; nilainya mengecil tapi peringkat Anda di kelas tidak berubah.
* **Cara Membaca Rumus:**

$$X_{scaled} = \frac{X_i - X_{min}}{X_{max} - X_{min}}$$



*Cara gampang membacanya:* "Kurangi nilai asli dengan nilai terendah, lalu bagi dengan selisih antara nilai tertinggi dan terendah (jarak total data)."
* **Penjelasan Kode Python:**
```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()                  # Membuat objek cetakan untuk Min-Max Scaling
scaled_data = scaler.fit_transform(df)   # Mempelajari min-max data sekaligus mengubah nilainya menjadi skala 0-1

```



### 🧭 3. Normalization (Vector Normalization)

* **Penjelasan:** Teknik ini bekerja secara horizontal (per baris/sampel), bukan per kolom. Ia mengubah data sehingga panjang vektor (Euclidean norm) dari setiap baris menjadi bernilai $1$. Fokusnya adalah pada arah data, bukan pada besar/magnitudo nilainya.
* **Analogi/Contoh:** Seperti menyamakan panjang semua anak panah menjadi 1 meter tanpa mengubah arah bidikan anak panah tersebut.
* **Cara Membaca Rumus:**

$$X_{scaled} = \frac{X_i}{\|X\|}$$



*Cara gampang membacanya:* "Nilai tiap kolom di satu baris dibagi dengan total panjang jarak (Euclidean) dari baris tersebut."
* **Penjelasan Kode Python:**
```python
from sklearn.preprocessing import Normalizer
scaler = Normalizer()                    # Membuat objek pengubah untuk Vector Normalization
scaled_data = scaler.fit_transform(df)   # Mengubah setiap baris data agar memiliki panjang unit yang sama

```



### ⚖️ 4. Standardization (Z-Score Normalization)

* **Penjelasan:** Mengubah data agar berpusat di angka $0$ (rata-rata/mean menjadi $0$) dan memiliki simpangan baku (standard deviation) bernilai $1$. Sangat optimal digunakan untuk algoritma yang mengasumsikan data berdistribusi normal (seperti Linear Regression atau SVM).
* **Analogi/Contoh:** Menilai performa murid bukan dari nilai mentahnya, melainkan dari seberapa jauh nilai murid tersebut berada di atas atau di bawah rata-rata nilai kelas.
* **Cara Membaca Rumus:**

$$X_{scaled} = \frac{X_i - \mu}{\sigma}$$



*Cara gampang membacanya:* "Nilai asli dikurangi rata-rata seluruh data ($\mu$), lalu hasilnya dibagi dengan nilai standar deviasinya ($\sigma$)."
* **Penjelasan Kode Python:**
```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()                # Membuat objek untuk standarisasi Z-score
scaled_data = scaler.fit_transform(df)   # Menghitung mean & standar deviasi lalu mengubah data agar berpusat di 0

```



### 🛡️ 5. Robust Scaling

* **Penjelasan:** Teknik scaling yang paling aman dari gangguan data rusak atau pencilan (*outliers*). Alih-alih memakai rata-rata, teknik ini menggunakan nilai tengah (median) dan rentang interkuartil (IQR) untuk membagi data.
* **Analogi/Contoh:** Jika Anda menghitung rata-rata kekayaan warga di sebuah desa dan tiba-tiba ada seorang miliarder pindah ke desa itu, angka rata-rata akan rusak. Namun, nilai tengah (median) toko desa akan tetap stabil.
* **Cara Membaca Rumus:**

$$X_{scaled} = \frac{X_i - X_{median}}{IQR}$$



*Cara gampang membacanya:* "Nilai asli dikurangi nilai tengahnya, lalu dibagi dengan jarak antara kuartil atas (persentil 75) dan kuartil bawah (persentil 25)."
* **Penjelasan Kode Python:**
```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()                  # Membuat objek scaling yang kebal terhadap pencilan
scaled_data = scaler.fit_transform(df)   # Mentransformasikan data menggunakan median dan rentang IQR

```

## 🛠️ Kesimpulan: Alur Kerja Pemilihan Scaling

Untuk mempermudah eksperimen di Jupyter Notebook, gunakan aturan alur logika ($Workflow$) berikut sebagai acuan cepat Anda saat melakukan data preprocessing:

$$\text{Mulai Preprocessing} \longrightarrow \text{Cek Outliers/Pencilan Data}$$

$$\Downarrow$$

$$\begin{cases} 
\text{Banyak Outliers Ekstrem} & \Longrightarrow \boxed{\text{Robust Scaling}} \\ 
\text{Data Bersih / Distribusi Normal} & \Longrightarrow \boxed{\text{Standardization (Z-Score)}} \\
\text{Butuh Batasan Nilai Pasti (0 sampai 1)} & \Longrightarrow \boxed{\text{Min-Max Scaling}} \\
\text{Analisis Teks / Kemiripan Arah Vektor} & \Longrightarrow \boxed{\text{Vector Normalization}} 
\end{cases}$$
