```python
import pandas as pd
```

## Membaca CSV di Python dengan Pandas

```python
file = "../0.Dataset/people_data.csv"
df = pd.read_csv(file)
df
```

### Fitur-fitur dalam Pandas read_csv

### 1. Baca kolom tertentu menggunakan read_csv

```python
df = pd.read_csv(file,usecols=['First Name','Email'])
df
```

### 2. Mengatur Kolom Indeks (index_col)

```python
df = pd.read_csv(file,index_col='First Name')
df
```

### 3. Menangani Nilai yang Hilang Menggunakan read_cs

```python
df = pd.read_csv(file,na_values=["N/A"])
df # gk bisa
```

### 4. Membaca File CSV dengan Pembatas yang Berbeda

```python
df = pd.read_csv('../0.Dataset/sample_bill.csv',sep='[:,|_]',engine='python')
df
```

### 5. membatasi jumlah baris yang dibaca

```python
df = pd.read_csv(file,nrows=3)
df
```

### 6. melewati baris yang tidak perlu di awal file,

```python
df = pd.read_csv(file)
print(f"Previous Dataset:\n {df}")

df = pd.read_csv(file,skiprows=[4,5])
print(f"Datasrt Sfter skipping  rows:\n {df}")
```

### 7. Mengurai Tanggal (parse_dates)

```python
df = pd.read_csv(file,parse_dates=['Date of birth'])
df.info()
```
