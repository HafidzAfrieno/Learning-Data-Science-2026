```python
import numpy as np
```

### Membuat Array Numpy

```python
arr = np.array([1,2,3])
print(arr)

arr = np.array([3,5,2,7])
print(arr)

arr = np.array([9,2,4,3])
print(arr)
```

### Mengakses Indeks Array

```python
arr = np.array([[-1, 2, 0, 4],
                [4, -0.5, 6, 0],
                [2.6, 0, 7, 8],
                [3, -7, 4, 2.0]])

arr2 =arr[:2,::2]
print(f"2 baris pertama dan kolom berselanh-seling (0,2):\n {arr2}")

arr3 = arr[[1, 1, 0, 3], 
            [3, 2, 1, 0]]
print(f"\nElemen pada indeks (1,3),(1,2),(0,1),(3,0): {arr3}")
```

### Operasi Array Dasar

```python
a = np.array([[1,2],
              [3,4]])
b = np.array([[4,5],
              [2,1]])

print(f"Menambah 1 ke setiap element A:\n {a + 1}")
print(f"\n Mengurangi 2 dari setia elemen B:\n {(b-2)}")
print(f"\n Jumlah semua elemen array:\n {a.sum()}")
print(f"\nJumlah array:\n {a+b}")
```

### Membangun Objek Tipe Data

```python
x = np.array([1,2])
print(f"{x} bertipe data: {x.dtype}")

x = np.array([0.1,0.2])
print(f"{x} bertipe data: {x.dtype}")

x = np.array([1, 2], dtype = np.int64) 
print(f"{x} bertipe data: {x.dtype}")
```

### Operasi Matematika pada Tipe Data array


```python
arr1 = np.array([[4,7],
                 [2,6]])

arr2 = np.array([[3,6],
                 [2,8]])

sum = np.add(arr1,arr2)
print(f"function add:\n {sum}")

sum1 = np.sum(arr1)
print(f"function sum: {sum1}")

sqrt = np.sqrt(arr1)
print(f"function sqrt:\n {sqrt}")
```
