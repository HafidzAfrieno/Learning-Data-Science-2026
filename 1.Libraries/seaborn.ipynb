```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

## Creating plots with Seaborn

### 1.Line Plot

```python
data = {'Name': ['ANSH', 'SAHIL', 'JAYAN', 'ANURAG'], 'Age': [21, 23, 20, 24]}
df = pd.DataFrame(data)

sns.lineplot(x=df.index,y='Age',data=df)
plt.xlabel('Index')
plt.ylabel('Age')
plt.title('Age Line Plot')
plt.show()
```

### 2.Scatter Plot

```python
data = {'Name': ['ANSH', 'SAHIL', 'JAYAN', 'ANURAG'], 'Age': [30, 23, 20, 24]}
df = pd.DataFrame(data)

sns.scatterplot(data=df,x='Name',y='Age')
plt.show()
```

### 3.Box plot

```python
data = {'Name': ['ANSH', 'SAHIL', 'JAYAN', 'ANURAG'], 'Age': [10, 3, 20, 140]} 
df = pd.DataFrame(data)

sns.boxplot(data=df,y='Age')
plt.show()
```

### 4.Violin Plot

```python
data = {'Name': ['ANSH', 'SAHIL', 'JAYAN', 'ANURAG'], 'Age': [21, 23, 20, 24]}
df = pd.DataFrame(data)

sns.violinplot(y='Age',data=df)
plt.show()
```

### 5. Swarm plot

```python
data = {'Name': ['ANSH', 'SAHIL', 'JAYAN', 'ANURAG'], 'Age': [21, 23, 20, 24]}
df = pd.DataFrame(data)

sns.swarmplot(x='Name', y='Age', data=df)
plt.show()
```

### 6. Bar Plot

```python
data = {'Name': ['ANSH', 'SAHIL', 'JAYAN', 'ANURAG'], 'Age': [21, 23, 20, 24]}
df = pd.DataFrame(data)

sns.barplot(data=df,x='Name',y='Age')
plt.show()
```

### 7. Point Plot

```python
data = {'Name': ['ANSH', 'SAHIL', 'JAYAN', 'ANURAG'], 'Age': [21, 23, 20, 24]}
df = pd.DataFrame(data)

sns.pointplot(data=df,x='Name',y='Age')
plt.show()
```

### 8. Count Plot

```python
data = {'Name': ['ANSH', 'SAHIL', 'ANSH', 'JAYAN', 'ANURAG', 'ANURAG', 'ANURAG', 'SAHIL']}
df = pd.DataFrame(data)

sns.countplot(x='Name',data=df)
plt.title('Frequensi')
plt.show()
```

### 9. Plot KDE
digunakan untuk memvisualisasikan Kepadatan Probabilitas dari variabel kontinu

```python
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df['species'] = iris.target
df['species']=df['species'].map({0:'Setosa',1:'Vercolor',2:'Virginica'})

sns.kdeplot(df[df['species']=='Virginica'],x='sepal length (cm)',label='Virginica',fill=True,color='red')
plt.legend()
plt.show()
```

### Cara Mengkustomisasi Plot Seaborn dengan Python

### 1. Menambahkan Judul dan Label Sumbu

```python
iris = sns.load_dataset('iris')
sns.scatterplot(x='sepal_length', y='sepal_width', data=iris)

# Add plot title and axis labels
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

### 2. Gaya dan Kisi Bawaan di Seaborn

```python
##stylegird
sns.set_style('whitegrid')

sns.boxplot(x='species', y='petal_length', data=sns.load_dataset('iris'))
plt.title('Petal Length Distribution by Species')
plt.show()
```

### 3. Menyesuaikan Palet Warna

```python
# a) Menggunakan Palet Bawaan:
sns.set_palette("pastel") 

sns.violinplot(x='species', y='petal_length', data=sns.load_dataset('iris'))
plt.title('Petal Length Distribution by Species')
plt.show()
```

```python
# b) Menggunakan Palet Kustom
custom_colors = ['#FF5733', '#33FFBD', '#335BFF']

sns.violinplot(x='species', y='petal_length', data=sns.load_dataset('iris'),palette=custom_colors,hue='species')
plt.title('Custom Colored Petal Length Distribution')
plt.show()
```

### 4. Menyesuaikan Ukuran Gambar dan Rasio Aspek

```python
plt.figure(figsize=(10, 6))

sns.lineplot(x='year', y='passengers', data=sns.load_dataset('flights'))
plt.title('Number of Passengers Over Time')
plt.show()
```

### 5. Menambahkan Penanda pada Grafik Garis

```python
sns.lineplot(x='year',y='passengers',data=sns.load_dataset('flights'),marker='o')
plt.title('Number of Passengers Over Time')
plt.show()
```

## Memvisualisasikan Hubungan dan Pola dengan Seaborn

### 1.Plot Pasangan

```python
sns.set_style('whitegrid')
palette = sns.color_palette('husl',8)

data= sns.load_dataset('iris')
sns.pairplot(data,hue='species')
plt.show()
```

### 2.plot Gabungan

```python
data = sns.load_dataset('tips')
sns.jointplot(x='total_bill',y='tip',data=data,kind='kde',color='red')
plt.show()
```

### 3.Grid Plot

```python
tips = sns.load_dataset('tips')

plot = sns.FacetGrid(tips,col='time',row='sex')
plot.map(sns.scatterplot,'total_bill','tip')
plt.show()
```

## Regression Plots: Visualizing Linear Relationships

```python
tips = sns.load_dataset('tips')

# regplot
sns.regplot(x='total_bill',y='tip',data=tips,scatter_kws={'s':10},line_kws={'color':'red'})
plt.show()
```

```python
#lmplot
sns.lmplot(x='total_bill',y='tip',data=tips)
plt.show()
```
