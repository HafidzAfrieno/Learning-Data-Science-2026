```python
import matplotlib.pyplot as plt
```

### 1. Grafik Garis

```python
x = [10,20,30,40]
y = [20,25,10,55]

plt.plot(x,y)
plt.title('Line Chart')
plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')
plt.show()
```

### 2. Bar Chart

```python
x=['akmal','Budi','Santoso','Edi']
y=[190,130,350,170]

plt.bar(x,y)
plt.title('Bar Chart')
plt.xlabel('Day')
plt.ylabel('total Point')
plt.show()
```

### 3. Histogram

```python
x = [7, 8, 9, 10, 10, 12, 12, 12, 13, 15, 14, 15, 16, 16, 17, 18, 18, 19, 20, 20,
     21, 22, 23, 24, 25, 25, 26, 28, 30, 32, 35, 36, 38, 40, 42, 44, 48, 50]

plt.hist(x,bins=10,color='steelblue')
plt.title('Histogram')
plt.xlabel('total Bill')
plt.ylabel('Frequnsi')
plt.show()
```

### 4. Scatter Plot

```python
x = ['Thur', 'Fri', 'Sat', 'Sun', 'Thur', 'Fri', 'Sat', 'Sun']
y = [170, 120, 250, 190, 160, 130, 140, 200]

plt.scatter(x,y)
plt.title('Scatter Plot')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

### 5. Pie Chart

```python
cars = ['AUDI', 'BMW', 'FORD','TESLA', 'JAGUAR',]
data = [23, 100, 35, 15, 12]

plt.pie(data,labels=cars)
plt.title('Pie Chart')
plt.show()
```

### 6. BoxPlot

```python
data = [ [10, 12, 14, 15, 18, 20, 22],
         [8, 9, 11, 13, 17, 19, 21],
         [14, 16, 18, 20, 100, 25, 27] ]

plt.boxplot(data)
plt.xlabel('Groups')
plt.ylabel('Values')
plt.title('Boxplot')
plt.show()
```

### 7. Heatmap

```python
import numpy as np
np.random.seed(0)
data = np.random.rand(10,10)

plt.imshow(data,cmap='viridis',interpolation='nearest')
plt.colorbar()
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Heatmap')
plt.show()
```

## Cara Mengkustomisasi Visualisasi Matplotlib

### 1. Menyesuaikan Grafik Garis

```python
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]
plt.plot(x,y,color='red',linewidth=3,marker='o',markersize=15,linestyle='--')

plt.title("Customizing Line Chart")
plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')
plt.show()
```

### 2. Mengkustomisasi Barplot

```python
x = ['Thur', 'Fri', 'Sat', 'Sun']
y = [170, 120, 250, 190]

plt.bar(x, y, color='red', edgecolor='black', linewidth=3)

plt.title("Customizing Bar Chart")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

### 3. Mengkustomisasi Plot Histogram

```python
x = [7, 8, 9, 10, 10, 12, 12, 12, 13, 14, 14, 15, 16, 16, 17,
     18, 18, 19, 20, 20, 21, 22, 23, 24, 25, 25, 26, 28, 30,
     32, 35, 36, 38, 40, 42, 44, 48, 50]

plt.hist(x, bins=10, color='red', edgecolor='black',linestyle='--', alpha=0.5)

plt.title("Customizing Histogram Plot")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

### 4. Mengkustomisasi Grafik Sebar

```python
x = ['Thur', 'Fri', 'Sat', 'Sun', 'Thur', 'Fri', 'Sat', 'Sun']
y = [170, 120, 250, 190, 180, 130, 260, 200]
size = [2, 3, 4, 2, 3, 2, 4, 3]  
bill = [170, 120, 250, 10, 100, 130, 260, 200]  

plt.scatter(x, y, c=size, s=bill, marker='D', alpha=0.5)

plt.title("Customizing Scatter Plot")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

### 5. Customizing Pie Chart

```python
cars = ['AUDI', 'BMW', 'FORD','TESLA', 'JAGUAR',]
data = [23, 13, 35, 15, 12]
explode = [0.1, 0.5, 0, 0, 0]
colors = ( "orange", "cyan", "yellow","grey", "green",)

plt.pie(data,labels=cars,explode=explode,autopct='%1.2f%%',colors=colors)
plt.show()
```

## Komponen Inti Matplotlib: Gambar dan Sumbu

### 1.Figure Class

```python
fig = plt.figure(figsize=(6,4),facecolor='lightblue') #figure class

ax = fig.add_axes([0.1,0.1,0.8,0.8])
x= [1,2,3,4]
y=[10,30,20,25]
ax.plot(x,y)

plt.title('Simple Line Plot')
plt.xlabel('X-Axes')
plt.ylabel('Y-Axis')
plt.show()
```

### 2. Axes Class

```python
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0.2,0.1,0.8,0.8]) #axes([left, bottom, width, height])

ax.plot(x,y)
ax.plot(y,x)

ax.set_title('Linear Graph')
ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')
ax.legend(labels=('Line 1','Line 2'))
plt.show()
```

## Teknik Lanjutan untuk Memvisualisasikan Subplot

### 1. Menggunakan add_axes()

```python
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

fig = plt.figure(figsize=(10,5))
ax1 = fig.add_axes([0.1,0.2,0.35,0.8])
ax2 = fig.add_axes([0.55,0.2,0.40,0.8])

ax1.plot(x,y)
ax2.plot(y,x)
plt.show()
```

### 2. Menggunakan subplot

```python
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

plt.figure()
plt.subplot(221)
plt.plot(x,y)

plt.subplot(212)
plt.plot(y,x)
plt.show()
```

### 3.Menggunakan sublot2grid

```python
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

axes1= plt.subplot2grid(
    (7,1),(0,0),rowspan=2,colspan=1
)

axes2=plt.subplot2grid(
    (7,1),(2,0),rowspan=2,colspan=3
)

axes1.plot(x,y)
axes2.plot(y,x)
plt.show()
```
