### 7. Menangani Baris dan Kolom dalam Pandas DataFrame

```python
# 1. Menambahkan Kolom Baru ke DataFrame :
df = pd.DataFrame({'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35]})

df['Salary'] = [50000,60000,90000]
print(df)
```

### 8. Menangani Data yang Hilang di Pandas

```python
# 1. Mengidentifikasi Data yang Hilang dengan Pandas
import numpy as np
df = pd.DataFrame({'Col1': [1, 2, np.nan],'Col2': [3, np.nan, np.nan]})

print(df.isnull())
```

```python
# 2. Mengisi Data yang Hilang:
df = df.fillna(0)
print(df.isnull())
```

```python
# 3. Menghapus Data yang Hilang dengan Pandas
data = {'Name': ['Alice', 'Bob', None], 'Age': [25, None, 35]}
df= pd.DataFrame(data)
df_dropped = df.dropna()

print(df_dropped)
```

### 9. Agregasi dan Pengelompokan dengan Pandas

```python
data = {'Category': ['A', 'B', 'A', 'B'], 'Value': [10, 20, 30, 40]}
df = pd.DataFrame(data)

grouped_sum = df.groupby('Category').sum()
print(f"Sum:\n {grouped_sum}")

grouped_mean = df.groupby('Category').mean()
print(f"Mean:\n {grouped_mean}")

grouped_agg = df.groupby('Category').agg(['sum','mean'])
print(f"Aggregated:\n {grouped_agg}")
```
