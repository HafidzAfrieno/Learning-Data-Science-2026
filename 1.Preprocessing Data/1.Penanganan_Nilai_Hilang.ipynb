## 1.Menghapus Data dengan Nilai Hilang

### 1.Menghapus Baris dengan Nilai Hilang

```python
df_target = df.copy()
df_cleaned = df.dropna()

print(f'Jumlah baris sebelum: {df.shape[0]}')
print(f'Jumlah baris setelah: {df_cleaned.shape[0]}')
```

### 2.Menghapus Kolom dengan Nilai Hilang

```python
df_target = df.copy()
proporsi = len(df) * 0.5 # Menghapus kolom yang memiliki lebih dari 50% nilai hilang
df_cleaned = df.dropna(axis=1,thresh=proporsi)

print(f'Jumlah baris sebelum: {df.shape[0]}')
print(f'Jumlah baris setelah: {df_cleaned.shape[0]}')
```

## 2.Mengisi Nilai Hilang

```python
colom_numerik = df.select_dtypes(include=np.number).columns
colom_kategori = df.select_dtypes(include='object').columns
```

### 1.Mengisi dengan Mean

```python
df_target = df.copy()
imuter = df_target[colom_numerik].mean().round()
df_target[colom_numerik] = df_target[colom_numerik].fillna(imuter)

print(f'Imuter:\n{imuter}\n')
print(f'Jumlah baris sebelum:\n{df[colom_numerik].isnull().sum()}\n')
print(f'Jumlah baris setelah:\n{df_target[colom_numerik].isnull().sum()}')
```

### 2.Mengisi dengan Median

```python
df_target = df.copy()
imuter = df_target[colom_numerik].median().round()
df_target[colom_numerik] = df_target[colom_numerik].fillna(imuter)

print(f'Imuter:\n{imuter}\n')
print(f'Jumlah baris sebelum:\n{df[colom_numerik].isnull().sum()}\n')
print(f'Jumlah baris setelah:\n{df_target[colom_numerik].isnull().sum()}')
```

### 3.Mengisi dengan Mode

```python
df_target = df.copy()
imuter = df_target[colom_kategori].mode().iloc[0]
df_target[colom_kategori] = df_target[colom_kategori].fillna(imuter)

print(f'Imuter:\n{imuter}\n')
print(f'Jumlah baris sebelum:\n{df[colom_kategori].isnull().sum()}\n')
print(f'Jumlah baris setelah:\n{df_target[colom_kategori].isnull().sum()}')
```

### 4.Mengisi dengan Nilai Tetap

```python
df_target = df['Store_Location'].copy()
imuter = {'Store_Location':'Indonesia'}
df_target = df_target.fillna(imuter)

print(f'Imuter:{imuter}\n')
print(f'Jumlah baris sebelum: {df['Store_Location'].isnull().sum()}\n')
print(f'Jumlah baris setelah: {df_target.isnull().sum()}')
```

### 5.Mengisi dengan Forward Fill

```python
df_target = df.copy()
df_target = df_target.ffill()#mengisi nilai kosong dengan nilai beris sebelumya

print(f'Jumlah baris sebelum:\n{df.isnull().sum()}\n')
print(f'Jumlah baris setelah:\n{df_target.isnull().sum()}')
```

### 6.Mengisi dengan Backward Fill

```python
df_target = df.copy()
df_target = df_target.bfill() #mengisi nilai kosong dengan nilai beris sesudahnya

print(f'Jumlah baris sebelum:\n{df.isnull().sum()}\n')
print(f'Jumlah baris setelah:\n{df_target.isnull().sum()}')
```
